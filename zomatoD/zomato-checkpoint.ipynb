```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop(['url','phone','dish_liked','reviews_list','address','menu_item'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace =True)
df.shape
```

```python
# rate col have eg 4.1/5  or any  , want simple float value.
df['rate'].unique()
```

```python
# here two string values to ie , NEW, nan

def handle_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value =str(value).split('/')      #2.2/5--2.2
        value = value[0]
        return float(value)

df['rate']=df['rate'].apply(handle_rate)
df['rate'].head()

    
```

```python
# null values in rate 
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Costplates','listed_in(type)':'Type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df['Costplates'].unique()
```

```python
# remove , from col
def handlecomma(value):
    value =str(value)
    if ',' in value:
        value = value.replace(',','')
    else:
        return float(value)

df['Costplates']= df['Costplates'].apply(handlecomma)
df['Costplates'].unique()
```

```python
df.head()
```

```python
# cleaning rest_type col
df['rest_type'].value_counts()
```

```python
rest_type =df['rest_type'].value_counts(ascending =False)
rest_type
```

```python
# making the cluster of those of 1,2 etc (less number of)--rest type
rest_types_lessthan1000=rest_type[rest_type<1000]
rest_types_lessthan1000
```

```python
# making rest types less than 1000 in frequency as other
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'other'
    else:
        return value

df['rest_type'] =df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
# cluster for less returntant

location =df['location'].value_counts(ascending =False)
location_lessthan300 =location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] =df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head() 
```

```python
# cuisines col
cuisines =df['cuisines'].value_counts(ascending =False)

cuisines_lessthan100 =cuisines[cuisines<100]
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] =df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
